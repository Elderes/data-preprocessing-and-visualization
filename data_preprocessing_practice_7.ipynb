```python
# 4
def replace_symbols(value):
  if not isinstance(value, str):
    return value
  else:
    for old_value, new_value in symbols.items():
      value = value.replace(old_value, new_value)
    return value

def str_to_float(value):
  if not isinstance(value, str):
    return value
  else:
    for old_value, multiplier in numeric.items():
      return float(value.replace(old_value, '')) * multiplier

def data_clean(value):
  x = replace_symbols(value)
  return str_to_float(x)

symbols = {'<=': '', '<': '', '>': '', '>=': ''}
numeric = {'K': 1000}

adult = pd.read_csv('/content/drive/MyDrive/Datasets/pandas 4/adult.data', header=None)
print(adult.iloc[:, -1].apply(data_clean).mean()) # last column is about salary. values are not accurate.
```

```python
# 9
# https://www.geeksforgeeks.org/python-opencv-cv2-calchist-method/
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/drive/MyDrive/Datasets/rat.jpg')
colors = ('b', 'g', 'r')

for i, col in enumerate(colors):
  hist = cv2.calcHist([image], [i], None, [255], [0, 256])
  plt.plot(hist, color=col)
  plt.xlim([0, 256])
plt.show()
```

```python
# 10
import cv2
import numpy as np

image = cv2.imread('/content/drive/MyDrive/Datasets/rat.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_blur = cv2.GaussianBlur(image_gray, (3, 3), 0)

edges = cv2.Canny(image_blur, 150, 150)

red_edges = np.zeros_like(image)
red_edges[edges != 0] = [255, 0, 0]
result = cv2.addWeighted(image, 0.6, red_edges, 1, 0)
result
```

```python
# 11
image = cv2.imread('/content/drive/MyDrive/Datasets/brain.tif')
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
cl = clahe.apply(l)
limg = cv2.merge((cl, a, b))
final = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

cv2_imshow(image)
cv2_imshow(final)
```

```python
# 12
# need dir with raw images
import cv2
import os

path = ''

for filename in os.listdir(path):
  if filename.endswith('.raw'):
    image = cv2.imread(f'{path}{filename}') # check the /
    adjusted = cv2.convertScaleAbs(image, alpha=1.5, beta=0)
    cv2.imwrite(f'converted/{filename[:-4]}.png', adjusted)
```
